```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Reading in Dataframes
df1 = pd.read_excel('By Constituent $ 1000 +2018-2023 20230825.xlsx')
df2 = pd.read_excel('Full Donation data set 1.15.23-1.15.24 BH Donations.xls')
df3 = pd.read_csv('Gift List Sept 1_Dec_1 2023.csv')
df4 = pd.read_csv('Give BIG CSV for Bethlehem Haven as of 12-14-2023, 16_57 UTC.csv')
df5 = pd.read_excel('One Day Critical Needs Donors _08_01_23.xlsx')
```

```python
df1.head()
```

Cleaning Date Columns

```python
df1['Unnamed: 7'] = pd.to_datetime(df1['Unnamed: 7'], format='%m/%d/%Y')
```

```python
df2.head()
```

```python
df2['Gift Date'] = pd.to_datetime(df2['Gift Date'], format='%Y-%m-%d', errors='coerce')
```

```python
df2['Gift Date'].dropna(inplace=True)
```

```python
df3.head()
```

```python
df3['Date'] = pd.to_datetime(df3['Date'], format='%m/%d/%y')
```

```python
df4.head()
```

```python
df4['Donation Date (UTC)'] = pd.to_datetime(df4['Donation Date (UTC)'], format='%m/%d/%y')
```

```python
df5.head()
```

```python
df5['Transaction Date Time'] = pd.to_datetime(df5['Transaction Date Time'])
```

Splitting dataframes into desired columns

```python
df1 = df1[['Unnamed: 2', 'Unnamed: 5', 'Unnamed: 7', 'Unnamed: 6']]
```

```python
df2 = df2[['Preferred ZIP', 'Gift Amount', 'Gift Date', 'Gift Type']]
```

```python
df3 = df3[['Date', 'Amount']]
```

```python
df4 = df4[['Donation Date (UTC)', 'Intended Donation', 'Zipcode', 'Recurring?']]
```

```python
df5 = df5[['Transaction Date Time', 'Amount', 'Zip Code', 'Are you a first time donor to this organization?']]
```

Merging desired Columns

```python
big_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

# Create a new column in the big DataFrame by combining the date columns
big_df['merged_date'] = big_df['Unnamed: 7'].combine_first(big_df['Transaction Date Time']) \
                      .combine_first(big_df['Donation Date (UTC)']) \
                      .combine_first(big_df['Date']) \
                      .combine_first(big_df['Gift Date'])
```

```python
big_df = big_df.drop(['Unnamed: 7', 'Transaction Date Time', 'Donation Date (UTC)', 'Date', 'Gift Date'], axis=1)
```

```python
# Create a new column in the big DataFrame by combining the date columns
big_df['merged_zipcode'] = big_df['Unnamed: 2'].combine_first(big_df['Zipcode']) \
                      .combine_first(big_df['Zip Code']) \
                      .combine_first(big_df['Preferred ZIP'])
```

```python
big_df = big_df.drop(['Preferred ZIP', 'Zip Code', 'Zipcode', 'Unnamed: 2'], axis=1)
```

```python
# Create a new column in the big DataFrame by combining the date columns
big_df['merged_amount'] = big_df['Unnamed: 5'].combine_first(big_df['Gift Amount']) \
                      .combine_first(big_df['Amount']) \
                      .combine_first(big_df['Intended Donation'])\
                       .combine_first(big_df['Amount'])
```

```python
big_df.isna().sum()
```

Cleaning merged_amount column by standardizing dollar numbers

```python
big_df['merged_amount'] = big_df['merged_amount'].map(lambda x: float('nan') if pd.isna(x) else float(str(x).replace(',', '').replace('$', '')))

# Optionally, round the values to 2 decimal places
big_df['merged_amount'] = big_df['merged_amount'].round(2)
```

```python
big_df['merged_date'] = pd.to_datetime(big_df['merged_date'], format='%Y-%m-%d')
```

```python
big_df['merged_date']
```

Taking only first 5 digits of zip code (some supplied with 9 digits)

```python
big_df['merged_zipcode'] = big_df['merged_zipcode'].str.split('-').str[0]
```

```python
# Create a new column in the big DataFrame by combining the date columns
big_df['merged_recurring'] = big_df['Unnamed: 6'].combine_first(big_df['Recurring?']) \
                      .combine_first(big_df['Are you a first time donor to this organization?']) \
                      .combine_first(big_df['Gift Type'])
```

```python
big_df.drop(['Unnamed: 5','Unnamed: 6', 'Gift Amount','Gift Type',  'Amount', 'Intended Donation', 'Recurring?', 'Are you a first time donor to this organization?'],axis=1, inplace=True)
```

```python
#Sorting in reverse chronological order
big_df = big_df.sort_values(by='merged_date', ascending=False)
```

Adding a key for a unique identifier

```python
big_df['key'] = range(len(big_df))

big_df['key'] = big_df.reset_index().index
```

```python
big_df.to_csv('MERGED_DONATION_DATA.csv', index=False)
```
